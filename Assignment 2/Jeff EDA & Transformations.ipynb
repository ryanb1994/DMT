```python
# Modules
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
# File
df = pd.read_csv('training_set_VU_DM_2014.csv')
```

Plot of percentage missing values.
And plot of outliers.

```python
dfmv = df.isnull().sum() / len(df) * 100
```

```python
ax = dfmv.plot(kind='bar', figsize=(10,2))
ax.set_xlabel('Variables')
ax.set_ylabel('% Missing values')
```

```python
ax = df.price_usd.plot(kind='box',rot = 90, figsize=(16,10)) # After reducing outliers
ax.set_xlabel('Price_usd')
ax.set_ylabel('USD')
```

```python
# Separate time
df["date_time"] = pd.to_datetime(df["date_time"])
df["year"] = df["date_time"].dt.year
df["month"] = df["date_time"].dt.month
```

```python
# Remove outliers price_usd
top2point5 = df.price_usd.quantile(0.975)
low2point5 = df.price_usd.quantile(0.025)
df = df[df.price_usd < top2point5]
df = df[df.price_usd > low2point5]
```
